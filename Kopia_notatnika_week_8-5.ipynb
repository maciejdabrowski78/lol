```python
def normalize(df_col):
    return (df_col-df_col.min())/(df_col.max()-df_col.min())
```

```python
moj_df_norm = moj_df.copy()

moj_df_norm['obesity'] = normalize(moj_df['Death rate']) # zastap 'nazwa_kolumny' kolumna w Twoim dataframie, a nastepnie powtorz to dla pozostalych kolumn
moj_df_norm['Lifetime, death (%)'] = normalize(moj_df['Death rate'])
moj_df_norm['Life expectancy'] = normalize(moj_df['Death rate'])
moj_df_norm['Death rate'] = normalize(moj_df['Death rate'])
moj_df_norm['GDP'] = normalize(moj_df['Death rate'])
moj_df_norm['Agriculture'] = normalize(moj_df['Death rate'])
moj_df_norm['Net income'] = normalize(moj_df['Death rate'])
moj_df_norm['Access to electricity'] = normalize(moj_df['Death rate'])
moj_df_norm['Expected  schooling'] = normalize(moj_df['Death rate'])
moj_df_norm['Unemployment with education,'] = normalize(moj_df['Death rate'])
moj_df_norm['Share of youth not in education'] = normalize(moj_df['Death rate'])

moj_df_norm.columns
```

```python
def srednia_wazona(dataframe_do_liczenia, wagi):
    licznik = (dataframe_do_liczenia*wagi).sum(axis=1)
    mianownik = sum(wagi)
    return licznik/mianownik
```

```python
moje_wagi = [-0.04, -0.105, -0.05, 0.05, -0.01, 0.015, 0.015, 0.022, 0.15, 0.03, 0.01, 0.23, 0.2, 0.148, 0.04, 0.04, 0.44, 0.03, -0.015, -0.2, ] # wagi mozemy zapisac wewnatrz nawiasow kwadratowych - tworzymy w ten sposob liste, o kt√≥rej dowiecie siƒô wiƒôcej w dalszej czesci kursu
sum(moje_wagi)
```

```python
(moj_df_norm_do_liczenia,)
moj_df_norm['srednia_wazona'] = srednia_wazona(moj_df_norm_do_liczenia, moje_wagi)
moj_df_norm

 
```

```python
moj_df_norm_uszeregowane = moj_df_norm.sort_values('srednia_wazona', ascending=False)
moj_df_norm_uszeregowane
```
